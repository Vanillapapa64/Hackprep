```python
import pandas as pd 
df=pd.read_csv("/Users/navkiratsingh/Documents/coding/job_data_merged_1.csv")
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import numpy as np
```

```python
df.info()
```

```python
df['Type'].mode()
```

```python
df['Workplace']=df['Workplace'].fillna("On-site")
df['Department']=df['Workplace'].fillna("Engineering")
df['Location']=df['Location'].fillna("London, England, United Kingdom")
df['Type']=df['Type'].fillna("Full time")
df.info()
```

```python
df.head(30)
```

```python
le=LabelEncoder()
dum_location=le.fit_transform(df['Location'])
dum_location=pd.DataFrame(dum_location)
```

```python
dfloc=pd.DataFrame(dum_location)
dfloc
```

```python
df['Location']=dfloc
```

```python
df
```

```python
dep_dumm=le.fit_transform(df['Department'])
df_dept=pd.DataFrame(dep_dumm)
df['Department']=df_dept
```

```python
df.drop('Unnamed: 0',axis='columns',inplace=True)
```

```python
cat_dumm=le.fit_transform(df['Category'])
df_cat=pd.DataFrame(cat_dumm)
df['Category']=df_cat
```

```python
work_dumm=le.fit_transform(df['Workplace'])
df_work=pd.DataFrame(work_dumm)
df['Workplace']=df_work
```

```python
type_dumm=le.fit_transform(df['Type'])
df_type=pd.DataFrame(type_dumm)
df['Type']=df_type
```

```python
df.info()
```

```python
df
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('Type',axis='columns')
y=df['Type']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
clf1=GaussianNB()
clf2=MultinomialNB()
clf1.fit(x_train,y_train)
clf2.fit(x_train,y_train)
```

```python
clf1.score(x_test,y_test)
```

```python
clf2.score(x_test,y_test)
```
