```python
import pandas as pd 
from sklearn import svm 
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.model_selection import GridSearchCV
```

```python
df=pd.read_csv("train.csv")
```

```python
df.drop('Name',axis='columns',inplace=True)
df.drop('Ticket',axis='columns',inplace=True)
df.drop('Cabin',axis='columns',inplace=True)
df.drop('PassengerId',axis='columns',inplace=True)
```

```python
le=LabelEncoder()
df['Sex']=pd.DataFrame(le.fit_transform(df['Sex']))
df['Embarked']=pd.DataFrame(le.fit_transform(df['Embarked']))
df
```

```python
df['Age'] = df['Age'].astype(float)
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
x=df.drop('Survived',axis='columns')
y=df['Survived']
```

```python
model_params={
    'svc':{
        'model': svm.SVC(gamma='auto'),
        'param':{
            'C':[1,5,10],
            'kernel':['linear', 'poly' ]
        }
    },
    'RandomForestClassifier':{
        'model':RandomForestClassifier(),
        'param':{
            'n_estimators':[1,5,10],
            'criterion':["gini", "entropy"]
        }
    },
    'GaussianNB':{
        'model':GaussianNB(),
        'param':{}
    },
    'MultinomialNB':{
        'model':MultinomialNB(),
        'param':{}
    },
    'LogisticRegression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'param':{
            'C':[1,5,10]
        }
    },
    'decision_tree':{
        'model':DecisionTreeClassifier(),
        'param':{
            'criterion':['gini','entropy']
        }
    }
}
```

```python
model_params1 = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10]
        }
    },
    'naive_bayes_gaussian': {
        'model': GaussianNB(),
        'params': {}
    },
    'naive_bayes_multinomial': {
        'model': MultinomialNB(),
        'params': {}
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini','entropy'],
            
        }
    }     
}
```

```python
score=[]
for model_name,mp in model_params1.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(x,y)
    score.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_param': clf.best_params_
    })
```
